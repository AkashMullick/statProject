### Research Question

If you make claims that you are more honest before taking a test, does it impact the rate of cheating?


### Hypothesis

If the student taking the test fills out a survey making claims that they don’t cheat, it will reduce the rate of cheating compared to a student that doesn’t fill out the survey.

### Data Cleaning

First, we replace all of the blank cells in the "Survey or No" column with the words "No Survey". This will make it easier for us to work with the data in the future.

```python
data["Survey or No"].fillna("No Survey", inplace = True) 
# data
```

We create a new column called "Difference", which is equal to the difference between the number of answers reported correct and the actual number of answers correct for each person.

```python
data["Difference"] = data["Answers Reported Correct"] - data["Answers Correct"]
data = data[["Answers Correct", "Answers Reported Correct", "Difference", "Survey or No", "What does Honesty mean to you", "Definition #", "How important is honesty to you?", "Honesty in romantic relationship?", "Honesty academically?", "Subject's honesty?"]] # rearrange columns
# data
```

We use a technique called boolean indexing to select only the rows that correspond to people who took the honesty survey and copy them into a new DataFrame called "survey".

```python
survey = data[(data["Survey or No"] == "Survey") | (data["Survey or No"] == "survey")] # create a DataFrame with only the people who had the survey
# survey
```

We use boolean indexing again to select the people who did not take the honesty survey.

```python
noSurvey = data[(data["Survey or No"] == "No Survey")] # create a DataFrame with only the people who did not have the survey
noSurvey = pd.DataFrame(noSurvey, columns = ["Answers Correct", "Answers Reported Correct", "Difference", "Survey or No"])
# noSurvey
```

Using boolean indexing again, we create a DataFrame with only the people who both took the survey and cheated (reported that they answered more questions correctly than they actually did).

```python
surveyCheated = survey[(survey["Difference"] > 0)] # create a DataFrame with only the people who had the survey and cheated
surveyCheated
```

We also create a DataFrame with only the people who cheated but did not take the survey.

```python
noSurveyCheated = noSurvey[(noSurvey["Difference"] > 0)] # create a DataFrame with only the people who did not have the survey and cheated
noSurveyCheated
```

**Proportions**

The proportion of people in the sample who took the survey and cheated is calculated by dividing the number of people who took the survey and cheated by the total number of people who took the survey.

```python
proportionSurveyCheated = (len(surveyCheated.index))/(len(survey.index))
proportionSurveyCheated
```

The proportion of people in the sample who did not take the survey and cheated is calculated by dividing the number of people who did not take the survey and cheated by the total number of people who did not take the survey.

```python
proportionNoSurveyCheated = (len(noSurveyCheated.index))/(len(noSurvey.index))
proportionNoSurveyCheated
```

The pooled proportion of people who cheated is equal to the total number of people who cheated divided by the total number of people in the sample.

```python
pooledProportion = (len(surveyCheated.index) + len(noSurveyCheated.index))/(len(survey.index) + len(noSurvey.index))
pooledProportion
```

**Final Calculations**

Calculate the standard deviation of the difference between the proportion of students who didn't fill out a survey about their level of honesty and cheated and the proportion of students who did fill out the survey and cheated.

```python
import math
sd = math.sqrt((pooledProportion*(1-pooledProportion)/len(survey.index)) + (pooledProportion*(1-pooledProportion)/len(noSurvey.index)))
sd
```

Calculate the t-score that corresponds to the proportions.

```python
tScore = (proportionNoSurveyCheated - proportionSurveyCheated)/sd
tScore
```

Import and use the SciPy library to use the tcdf function to determine the p-value for the hypothsis test.

```python
from scipy import stats
pValue = 1 - stats.t.cdf(tScore,df = len(survey) - 1)
pValue
```
