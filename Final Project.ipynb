## Part 1: Focus

### Research Question

If you make claims that you are more honest before taking a test, does it impact the rate of cheating?


### Hypothesis

If the student taking the test fills out a survey making claims that they don’t cheat, it will reduce the rate of cheating compared to a student that doesn’t fill out the survey.


### Research

[Predictably Irrational Chapter 9](https://drive.google.com/file/d/1SESWl7V5kUJtoPihaG8Wy8XHnasDzwYM/view)

In this section, it talks about the priming effect on people’s behavior after viewing a stimulus. An example of the priming effect would be subjecting elderly individuals to words, and observing how their behavior reflects that stimulus. When the individuals left the building, their walking speed was noticeably slower than before they read the stimulus words.

[Results of a similar experiment](https://www.elsevier.com/editors-update/story/publishing-ethics/a-fascinating-experiment-into-measuring-dishonesty)

### Operational Definitions

The **independent variable** in this study is implementing a survey before the test with priming words linked to honesty and test integrity. The **dependent variable** that is being measured is the number of individuals that cheat on the test. **Cheating** in this experiment is reporting different amount of answers correct than actual correct answers.

## Part 4: Participants

### Participant Populations and Subgroups

The participant population is students from Carrboro High school in either honors or AP classes. The subgroups will be the group students who receive the survey and the group of people who do not receive the survey. 

### Representative Sample

We are sampling data from two AP classes and two honors math classes. By taking samples from classes of different difficulties, we are making sure the sample represents the population better than it would if we only sample one level of class.

### Random Assignment

To ensure random assignment, the groups chosen will be assigned random numbers and using a random number generator, we will determine which group is the experimental or the control.

## Part 5: Data Collection

### Description

The quiz will be given to each of the groups (control and experimental) and then once the answers are given they will self check. After they have graded their quiz, they will write their score on the paper. If they get all the questions right, they will get candy. The papers will be collected and we will determine whether they had changed their answers and if the score they gave themselves at the top of the page is accurate. This will be run on both groups, and we will compare the amount of cheating between the control and the experimental groups through google sheets.
